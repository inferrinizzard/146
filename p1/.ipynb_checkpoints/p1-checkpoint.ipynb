```python
from p1_support import load_level, show_level, save_level_costs
from math import inf, sqrt
from heapq import heappop, heappush
import p1
```

```python
graph = load_level("example.txt")
solution = load_level("example_a_to_e.txt")
graph
level = {**graph["spaces"],**{pos:1.0 for pos in graph["waypoints"].values()}}
```

```python
def path(start, end, graph, adj):
    q = [(0, start)] # node queue
    visited = {start:(0,None)} # dist, prev

    # while queue not empty
    while q:
        cur_dist, cur = heappop(q) # get current
        if cur == end: # check success
            path = [cur]
            while path[0] is not start: # backpathing
                path = [visited[path[0]][1]] + path
            return path
        for pos, cost in adj(graph, cur): # for each neighbour
            if cost is not inf:
                pathcost = cost + cur_dist
                if not visited.get(pos) or pathcost < visited[pos][0]:
                    visited[pos] = (pathcost,cur)
                    heappush(q, (pathcost, pos))
    return None
```

```python
start = (3,3)
level_cost = {}
for pos in level:
    p = path(start, pos, graph, p1.navigation_edges)
    cost = 0 if p else inf
    if p:
        for i in range(len(p)-1):
            a = p[i]
            b = p[i+1]
            cost = cost + (level[a] + level[b])/2*(1 if abs(a[0]-b[0]) + abs(a[1]-b[1]) >1 else sqrt(2))
    level_cost[pos] = cost
level_cost
```
