```python
from math import inf
from time import sleep
def path(source_point, destination_point, mesh):
    def find_box(point):
        return next((box for box in mesh["boxes"] if point[0] >= box[0] and point[0] < box[1] and point[1] >= box[2] and point[1] < box[3]), None)
    start = find_box(source_point)
    end = find_box(destination_point)
    if not start or not end:
        print("No path!")
        return [], []
    if start == end:
        return [(source_point, destination_point)], [start]

    grid = {}

    def centre(b):
        return ((b[0]+b[1])/2, (b[2]+b[3])/2)

    def Node(bounds, point=None, h=0, g=inf, f=inf, prev=None, head=True):
        grid[bounds] = {"pos": point if point else centre(
            bounds), "h": h, "g": g, "f": f, "prev": prev, "head":head}
        return bounds

    def dist(a, b):
        return (lambda c, d: ((d[0]-c[0])**2+(d[1]-c[1])**2)**.5)(centre(a), centre(b))
        # return (((b[0]+b[1])/2-(a[0]+a[1])/2)**2 + ((b[2]+b[3])/2-(a[2]+a[3])/2)**2)**.5

    def tuplify(boxes):
        centres = [centre(b) for b in boxes]
        return zip(centres[1:], centres[:-1])

    Node(start, source_point, g=0)
    Node(end, destination_point, head=False)

    print(start)
    print(end)

    open_set = [start]
    path = []

    while open_set:
        # get node with min f
        current = min(open_set, key=lambda n: grid[n]["f"])
        print(current)
        if(current == end):  # if at goal
            back = current
            while back:
                path.append(back)
                back = grid[back]["prev"]
            print(path)
            return tuplify(path), grid.keys()

        open_set.remove(current)
        for neighbour in mesh["adj"][current]:  # check each neighbour
            temp_g = grid[current]["g"] + dist(current, neighbour)
            # check predicted dist
            if neighbour not in grid.keys() or temp_g < grid[neighbour]["g"]:
                Node(neighbour, g=temp_g, f=temp_g +
                     dist(neighbour, end if grid[current]["head"] else start), prev=current, head=grid[current]["head"])
                if neighbour not in open_set:
                    open_set.append(neighbour)

    # tuple path
    return [], grid.keys()

path((680,888),(100,100),mesh_map)
```

```python
from math import inf
class Node():
        def __init__(self, point, bounds = None, h = 0, g = inf): # self, (float, float), float, float) → void
            self.bounds = bounds
            self.x = point[0]
            self.y = point[1]
            self.h = h
            self.g = g
            self.f = h + g
            self.prev = None

        # gets and sets heuristic based on euclidian dist
        def dist(self, other): # self, Node → float
            self.h = ((other.x-self.x)**2 + (other.y-self.y)**2)**.5
            return self.h

        # gets neighbours from mesh map
        def get_neighbours(self, boxes, mesh): # self, list of boxes, mesh of boxes
            adj = mesh["adj"][self.bounds]
            neighbours = []
            for box in adj:
                maybe_box = next((box == b.bounds for b in boxes), None)
                if not maybe_box:
                    neighbours.append(Node(((box[0]+box[1])/2, (box[2]+box[3])/2), box))
            return neighbours

    # to_string
        def __repr__(self):
            return "pos:{" + str(self.x) + "," + str(self.y) + "}" + ((" in " + str(self.bounds) + " ") if self.bounds else "") + "{g:" + str(self.g) + ", h:" + str(self.h) + ", f:" + str(self.f) + "}"
```

```python
def find_path(start, end, mesh):
    open_set = [start]
    closed_set = []
    path = []
    current = start
    start.g = 0
    start.dist(end)
    start.f = start.h
    print(start)
    print(end)

    while open_set:
        current = min(open_set, key=lambda n: n.f) # get node with min f
        if(current.pos == end.pos): # if at goal
            back = current
            while back:
                path.append(back)
                back = back.prev
            return path
        
        open_set.remove(current)
        for neighbour in current.get_neighbours(mesh): # check each neighbour
            temp_g = current.g + current.dist(neighbour)
            if temp_g < neighbour.g: # check predicted dist
                neighbour.prev = current
                neighbour.g = temp_g
                neighbour.f = temp_g + neighbour.dist(end)
                if neighbour not in open_set: # does not filter dupes
                    open_set.append(neighbour)
    return False
```
